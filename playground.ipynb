```python
import numpy as np
```

```python
def calculate_average_embedding(hps, embeddings_dict):
    embeddings = [embeddings_dict[hp_id] for hp_id in hps if hp_id in embeddings_dict]
    return np.mean(embeddings, axis=0) if embeddings else []
```
